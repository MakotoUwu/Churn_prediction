```python

# Identify potential outliers in the dataset for numerical features
outliers_data = {}

# Define a function to calculate IQR and identify outliers
def identify_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((col < lower_bound) | (col > upper_bound))

# Identify outliers for each numerical feature
for feature in numerical_features:
    outliers_data[feature] = identify_outliers(data_eda[feature])
```
