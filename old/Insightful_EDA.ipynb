
# Exploratory Data Analysis (EDA) for Churn Prediction
This notebook provides a step-by-step EDA for the bank churn prediction dataset.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv("../data/processed/BankChurners_clean.csv")
data.head()
```

```python

info = data.info()
description = data.describe()
description
```

```python

# Distribution of Customer Age
sns.histplot(data['Customer_Age'], kde=True)
plt.title('Distribution of Customer Age')
plt.show()

data['Gender'].value_counts().plot(kind='bar')
plt.title('Distribution of Gender')
plt.show()

```

```python

missing_values = data.isnull().sum()
missing_values
```

```python

# Select only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix for numeric columns
correlation_matrix = numeric_data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
